```python
# Now we will split_exp feature and perform exceptional handling
def split_exp2(exp):
  try:
    if len(exp.split('-'))==2:
      min_exp=exp.split('-')[0]
      max_exp=exp.split('-')[1]
    return pd.Series([min_exp,max_exp])
  except:
    return pd.Series([np.nan,np.nan]) 
```

```python
#Now we will define a  function which will check the dtype using exception handling
```

```python
def fetch_dt_att(dataframe,feature):
  try:
    return pd.Series([dataframe[feature].dt.day,dataframe[feature].dt.month,dataframe[feature].dt.year])
  except:
    print('Data type is not supported')  
```
